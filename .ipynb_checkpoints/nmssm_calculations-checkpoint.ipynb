# Try to reproduce the results using NMSSMCalcEW

- Run this to try to replicate figure 1 from 2206.04618. I will work with $\tan \beta = 3$ which means that $\beta \approx 1.24905$.
- I will further assume that $\mu_{eff} = 1.5TeV$ and that $m_A = 3 TeV$

- Using 0910.1785 -- equation 2.28 gives an expression for the mass of the charged Higgs eigenstate.
- We will use that $m_A^2 = \mathcal{M}^2_{P,11}$ and that $m_W = \frac{g_2 v}{2}$ and that $\hat{m_3} = 0$ for the $Z_3$ preserving NMSSM
$$M_{H^{\pm}}^2 = m_A^2 + 2 m_W^2 - v^2 \lambda^2$$

```python
os.chdir('/mnt/c/Users/asgia/Documents/research/nmssmcalcew/nmssmcalc_rew_alphat2-master/')
updater = NMSSMParamsUpdater('inpTrial.dat', 174)
```

```python
## Parameters 
mw = 8.03790000E+01   # Mw
beta = 1.24905 # β for tanβ=3
#Alam = ( 9e06 * np.sin(beta) * np.cos(beta)) / 1500 - 1500 
#v = 174 # SM Higgs vev
mueff = 1500 # effective mu 
mA = 3000 # mass of the pseudoscalar Higgs
```

```python
nmssm_params = pd.DataFrame(columns=['lam', 'kappa', 'mh', 'mHpm'])

for LAM in np.linspace(0.001,1,100):
    run_params = [LAM, LAM]
    dimensionful_param = updater.update_params(LAM,'27', mueff, mA, mw, beta)
    with open('output_trial_log_file.log', 'w') as f:
    # Run the command and redirect both stdout and stderr to the file
        subprocess.run('./run inptrial.dat outtrial.in decout_trial.out', stdout=f, stderr=f, shell=True)
    mass = updater.find_higgs_mass('outtrial.in')
    run_params.append(mass)
    run_params.append(dimensionful_param)
    nmssm_params.loc[len(nmssm_params)] = run_params


    
```

```python
nmssm_params.head()
```

```python
fig1 = pd.read_csv('/mnt/c/Users/asgia/Documents/research/nmssmcalcew/2206p04618_fig1_left_panel.csv')
fig1.head(2)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(fig1['lambda'], fig1['mh'] , label=r'2206.04618')
plt.plot(nmssm_params['lam'], nmssm_params['mh'], label=r'my version')
plt.xlabel(r'$\lambda$')
plt.ylabel(r'$m_h$ (GeV)')
plt.title(r'NMSSM higgs mass prediction for $tan \beta = 3$')
# plt.yscale('log')
# plt.xscale('log')
plt.legend()
plt.grid(True)
plt.show()
```

```python
M_values = np.unique(heatmapdf['M'])

# Create a dictionary with interpolated data for each M value
data_dict = {M: {
    'ratio_values': heatmapdf[heatmapdf['M'] == M]['Xt/M'].values,
    'lam_values': heatmapdf[heatmapdf['M'] == M]['lambda'].values,
    'kappa_values': heatmapdf[heatmapdf['M'] == M]['kappa'].values
} for M in M_values}

# Define the grid for interpolation
grid_lambda = np.linspace(heatmapdf['lambda'].min(), heatmapdf['lambda'].max(), 500)
grid_kappa = np.linspace(heatmapdf['kappa'].min(), heatmapdf['kappa'].max(), 500)
grid_lambda, grid_kappa = np.meshgrid(grid_lambda, grid_kappa)

fig, axes = plt.subplots(1, 4, figsize=(18, 6), sharex=True, sharey=True, constrained_layout=True)

for i, M in enumerate([1000, 2000, 4000, 5000]):
    # Get values for current M
    lam_values = data_dict[M]['lam_values']
    kappa_values = data_dict[M]['kappa_values']
    ratio_values = data_dict[M]['ratio_values']

    # Perform grid interpolation
    grid_ratio = griddata(
        (lam_values, kappa_values), ratio_values, 
        (grid_lambda, grid_kappa), method='linear'
    )

    # Plot the interpolated data as a filled contour plot
    im = axes[i].contourf(grid_lambda, grid_kappa, grid_ratio, cmap='plasma')
    axes[i].set_title(f'Heatmap of $X_t / M$ for $M = {M}$ GeV', fontsize=14)
    axes[i].set_xlabel(r'$\lambda$', fontsize=12)
    if i == 0:
        axes[i].set_ylabel(r'$\kappa$', fontsize=12)

# Add a single colorbar for all plots
fig.colorbar(im, ax=axes, location='right', label=r'$X_t/M$')

plt.show()
```
