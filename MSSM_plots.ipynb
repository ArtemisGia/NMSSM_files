## Import the results from SUSYHD and plot $m_h$ vs $X_t/M$

```python
mhSUSYHD = pd.read_csv('/mnt/c/Users/asgia/Documents/research/SUSYHD/3sigma_ILC500_Xt_Mh_mine.csv')
mhSUSYHD.head(2)
```

```python
# Plot
plt.figure(figsize=(8, 6))
plt.scatter(mhSUSYHD['XtMpos'], mhSUSYHD['Mhpos'], color='mediumblue', s=5)
plt.scatter(mhSUSYHD['XtMneg'], mhSUSYHD['Mhneg'], color='mediumblue', s=5)
plt.xlabel(r'$X_t / M $', fontsize=18)
plt.ylabel(r'$m_h$(GeV)', fontsize=18)
plt.axhline(y=125, color='gray', linestyle=':')
plt.grid(alpha=0.7)
plt.tight_layout()
plt.yscale('log')

text2 = r'$\tan \beta=20$'
text =  text2 
plt.text(-9.5, 5, text, fontsize=10, bbox=dict(facecolor='white', alpha=1))

plt.savefig("/home/venus/NMSSM_files/figures/MSSM_ILC500_3sigma_mh_vs_XtM.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
